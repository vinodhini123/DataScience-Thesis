```python
from scipy.spatial import distance
```

```python
def calc_eculidean_distance(m1,m2):
    distances=[]
    for i in m1:
        #ind=np.where(np.all(m1==i,axis=1))
        distances.append(distance.euclidean(i, m2))
    distances=np.vstack(distances)
    return distances
```

```python
def tricubic(x):
    y = np.zeros_like(x)
    idx = (x >= -1) & (x <= 1)
    y[idx] = np.power(1.0 - np.power(np.abs(x[idx]), 3), 3)
    return y
```

```python
def get_weights(distances, min_range):
	#print(distances[min_range])
    max_distance = np.max(distances[min_range])
    #print("max_distance ",max_distance )
    weights = tricubic(distances[min_range] / max_distance)
    return weights
```

```python
def get_min_range(distances, window):
        #print("distances",distances)
        min_idx = np.argmin(distances)
        n = len(distances)
        if min_idx == 0:
            return np.arange(0, window)
        if min_idx == n-1:
            return np.arange(n - window, n)

        min_range = [min_idx]
        #print("min_range in get_nim_range",min_range)
        while len(min_range) < window:
            i0 = min_range[0]
            i1 = min_range[-1]
            if i0 == 0:
                min_range.append(i1 + 1)
            elif i1 == n-1:
                min_range.insert(0, i0 - 1)
            elif distances[i0-1] < distances[i1+1]:
                min_range.insert(0, i0 - 1)
            else:
                min_range.append(i1 + 1)
        return np.array(min_range)
```

```python
pinf = float('+inf')
x_scale[x_scale == pinf] = 0.0
x_scale[np.isnan(x_scale)] = 0.0

X_train=X_train.to_numpy()
X_train[X_train == pinf] = 0.0
X_train[np.isnan(X_train)] = 0.0
```

```python
x_train_orig=x_train_orig.to_numpy()
```

```python
n_xx= x_scale
n_yy= y_scale
window =2
degree=1
y_pred=[]
for i in x_scale:
    n_x = i
    n_x[np.isnan(n_x)] = 0.0
    distances = calc_eculidean_distance(n_xx, n_x)
    min_range = get_min_range(distances, window)
    weights = get_weights(distances, min_range)
    wm = np.multiply(np.eye(window), weights)
    xm = np.ones((window, (n_xx.shape[1] * (degree)) +1))
    xp = np.array([[np.power(n_x, p)] for p in range(degree + 1)])
    xp=np.delete(xp, 0,0)
    xp=xp[0]
    xp=np.concatenate((np.array([1])[:, np.newaxis], xp), axis=1)
    #for i in range(1, degree + 1):
    xm[:, 1:n_xx.shape[1]+1] = np.power(n_xx[min_range], 1)
    ym=n_yy[min_range]
    xmt_wm = np.transpose(xm) @ wm
    beta = np.linalg.pinv(xmt_wm @ xm) @ xmt_wm @ ym
    y_calc = ( xp @ beta)[0]
    y_pred.append(y_calc)
y_pred=scaler_y.inverse_transform(y_pred)
```

```python
from numpy import sqrt
from sklearn.metrics import mean_squared_error
loess_RMSE=sqrt(mean_squared_error(y_train_orig,y_pred))
loess_RMSE
```
